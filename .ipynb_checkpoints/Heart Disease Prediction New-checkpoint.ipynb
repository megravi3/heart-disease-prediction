# PERFORMING NORMALIZATION AND CALCULATING ACCURACY

```python
Y = dataset.target.values
X_data = dataset.drop(['target'], 1)

# Min-Max Normalization
X = (X_data - np.min(X_data)) / (np.max(X_data) - np.min(X_data)).values

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,random_state=0)

#transpose matrices
X_train = X_train.T
Y_train = Y_train.T
X_test = X_test.T
Y_test = Y_test.T

from sklearn.linear_model import LogisticRegression
accuracies = {}
lr = LogisticRegression()
lr.fit(X_train.T,Y_train.T)
accuracy = lr.score(X_test.T,Y_test.T)*100

accuracies['Logistic Regression'] = accuracy
print(" Accuracy of our Regression Model is  {:.2f}%".format(accuracy))
```

```python
# Random Forest Regression

from sklearn.ensemble import RandomForestRegressor

rfY = dataset.target.values
rfX_data = dataset.drop(['target'], 1)

# Min-Max Normalization
rfX = (rfX_data - np.min(rfX_data)) / (np.max(rfX_data) - np.min(rfX_data)).values

rfX_train, rfX_test, rfY_train, rfY_test = train_test_split(rfX,rfY,test_size = 0.2,random_state=0)

#transpose matrices
rfX_train = rfX_train.T
rfY_train = rfY_train.T
rfX_test = rfX_test.T
rfY_test = rfY_test.T

#random forest for 1000 decision trees
rf = RandomForestRegressor(n_estimators = 500, random_state = 0)
rf.fit(rfX_train.T, rfY_train.T)

# Make prediction
predictions = rf.predict(rfX_test.T)

# Error
error = abs(predictions - rfY_test.T)
print('Mean Absolute Error:', round(np.mean(error), 2), 'degrees.')

# Calculate mean absolute percentage error (MAPE)
mape = 100 * (error / rfY_test.T)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
# Random Forest Regression

from sklearn.ensemble import RandomForestRegressor

rfY = np.array(dataset['target'])
rfX_data = dataset.drop(['target'], 1)
rfX_data1 = np.array(rfX_data)


rfX_train, rfX_test, rfY_train, rfY_test = train_test_split(rfX_data1,rfY,test_size = 0.2,random_state=0)

#transpose matrices
rfX_train = rfX_train.T
rfY_train = rfY_train.T
rfX_test = rfX_test.T
rfY_test = rfY_test.T

#random forest for 1000 decision trees
rf = RandomForestRegressor(n_estimators = 500, random_state = 0)
rf.fit(rfX_train.T, rfY_train.T)

# Make prediction
predictions = rf.predict(rfX_test.T)

# Error
error = abs(predictions - rfY_test.T)
print('Mean Absolute Error:', round(np.mean(error), 2), 'degrees.')

# Calculate mean absolute percentage error (MAPE)
mape = 100 * (error / rfY_test.T)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
